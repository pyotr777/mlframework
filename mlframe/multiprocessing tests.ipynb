```python
import multiprocessing
from Queue import Queue, Empty
#import subprocess
import os
import time
import re
import sys
```

```python
def worker(m, n, out_q):    
    p = multiprocessing.current_process()
    print "In",p.name,"nums=",nums
    output = p.name
    for i in range(n):
        output = m+n*i
        out_q.put(output)
        sys.stdout.flush()
        print "put",output
        time.sleep(1)
        
    print p.name,"finished"
    
    

# Each process will get 'chunksize' nums and a queue to put his out
# dict into
out_q = multiprocessing.Queue()
procs = []
nprocs = 3
nums = [4,5,6]

for i in range(nprocs):
    print "Starting",i
    p = multiprocessing.Process(name="W"+str(i),target=worker, args=(i,nums[i],out_q))
    procs.append(p)
    p.start()

result = ""
# Wait for all worker processes to finish
not_finished = nprocs
while not_finished > 0:
    # Read from queue
    try:
        out = out_q.get_nowait()
        print "Get:",out
        result += str(out)+","
    except Empty:
        print "-"
    
    # Check finished procs
    not_finished = nprocs
    for p in procs:
        if not p.is_alive():
            not_finished -= 1
    print "alive",not_finished,"procs"
    time.sleep(.5)

print "Result",result
```

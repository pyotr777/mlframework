```python
import multiprocessing
from Queue import Queue, Empty
#import subprocess
import os
import time
import re
import sys
import random
import pandas as pd
import time
```

```python
def worker(n, a, out_q):    
    p = multiprocessing.current_process()
    print "Start",a
    output = p.name
    max = 10
    gomi=["­ЪЇё","­ЪљА","­Ъљљ","­ЪЇ║","­Ъњђ","­ЪљЪ","­ЪљГ",
          "­Ъљ╣","­Ъљи","­Ъљѓ","­ЪљЉ","­ЪљЌ","­Ъљў","­Ъдњ","­ЪЇЁ",
          "­ЪЇє","­ЪЦЋ","­ЪјЃ","­Ъїй","­ЪЇћ","­ЪЇЎ","­ЪЇЋ","­ЪЇб","­ЪЉ╗"]
    gomi="qwertyuiopasdfghjkl"
    for i in range(n):
        num = random.randint(1,len(gomi))
        works=""
        for j in range(num):
            works += random.choice(gomi)
        output = a +"," + str(i)+","+works+"\n"
        start_time = time.time()
        if out_q.full():
            print "%",
        out_q.put(output)
        #sys.stdout.flush()
        elapsed_time = time.time() - start_time
        
        print " {:6.4f}s {}".format(elapsed_time,output),
        time.sleep(.01)
        
    print p.name,"finished"
    
    

# Each process will get 'chunksize' nums and a queue to put his out
# dict into
out_q = multiprocessing.Queue(5)
procs = []
names=["лљ","лЉ","лњ","лЊ","лћ","лЋ","лЂ","лќ","лЌ","лў","лЎ","лџ","лЏ","лю"]
nprocs = len(names)

for i in range(nprocs):
    p = multiprocessing.Process(name=names[i],target=worker, args=(nprocs,names[i],out_q))
    procs.append(p)
    p.start()

result = ""
# Wait for all worker processes to finish
not_finished = nprocs
while not_finished > 0:
    # Read from queue
    try:
        out = out_q.get_nowait()
        print "­Ъце",out,
        result += str(out)
    except Empty:
        print "-"
    
    # Check finished procs
    not_finished = nprocs
    for p in procs:
        if not p.is_alive():
            not_finished -= 1
    print "w",not_finished,"procs"    
    time.sleep(.1)
    while not out_q.empty():
        try:
            out = out_q.get_nowait()
            print "­Ъце",out,
            result += str(out)
        except Empty:
            print "!!!!"

#print "Result­Ъце\n",result
```
